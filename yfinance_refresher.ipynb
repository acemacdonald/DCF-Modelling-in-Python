What we really want if the aim is to build financial statements is to store these in separate datasets, this can be done by simply defining the ticker as a variable, then specifying which statement you are looking for.

```python
ticker = 'AAPL'

bal_sheet = yf.Ticker("AAPL").balance_sheet
inc_statement = yf.Ticker("AAPL").income_stmt
cf_statement = yf.Ticker("AAPL").cash_flow
```

Then if we want to look at our balance sheet in the latest year, we can quickly define the years by passing the amount of columns and then calling the function on the latest year.

```python
years = bal_sheet.columns

bal_sheet[years[0:1]]
```

Let's say we want to look at the last 3 years of the income statement, then we can similarly call the function to return our data.

```python
years = inc_statement.columns

inc_statement[years[0:3]]
```
